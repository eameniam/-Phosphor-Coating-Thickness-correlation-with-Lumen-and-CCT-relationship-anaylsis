# Experiment Result Visulization

In this notebook, I use experiment data to plot the comparison graph of lumen output, junction temperature and reliability test in 5000K and 10000K CCT.

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# load dataset of experiment data
df_5000 = pd.read_csv('5000K.csv')
df_10000 = pd.read_csv('10000KCCT.csv')
```

```python
# checking dataframe information
df_5000
```

```python
df_10000
```

```python
# loading the x and y value for plotting
x1=df_5000['thickness']
y1a=df_5000['Low Profile']
y1b=df_5000['Conventional']
```

```python
# loading the x and y value for plotting
x2=df_10000['Thickness']
y2a=df_10000['Low Profile']
y2b=df_10000['Conventional']
```

```python
# import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x1, y1a, label="Low Profile",color = "black", marker='o',markersize=12)
ax.plot(x1, y1b, label="Conventional",color = "black", marker='s',markersize=12)

ax.legend(loc=0,fontsize = 'x-large')

ax.set_title('(A)', y=-0.22,fontsize = 20)
ax.set_xlabel('Silicone Phosphor Mixture Coating Thickness(Î¼m)',fontsize = 16)
ax.set_ylabel('Lumen Output(lm)',fontsize = 16)

ax.errorbar(x1, y1a, yerr=1.5,color = 'black',fmt='o', capsize=10)
ax.errorbar(x1, y1b, yerr=1.5,color = 'black',fmt='o', capsize=10)

fig.savefig("5000lumen.png",bbox_inches='tight')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x2, y2a, label="Low Profile",color = "black", marker='o',markersize=12)
ax.plot(x2, y2b, label="Conventional",color = "black", marker='s',markersize=12)

ax.legend(loc=0,fontsize = 'x-large')

ax.set_title('(B)', y=-0.22,fontsize = 20)
ax.set_xlabel('Silicone Phosphor Mixture Coating Thickness(Î¼m)',fontsize = 16)
ax.set_ylabel('Lumen Output(lm)',fontsize = 16)

ax.errorbar(x2, y2a, yerr=2,color = 'black',fmt='o', capsize=10)
ax.errorbar(x2, y2b, yerr=2,color = 'black',fmt='o', capsize=10)

plt.savefig("10000lumen.png",bbox_inches='tight')
```
